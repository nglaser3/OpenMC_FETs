```python
import openmc
import numpy as np
from numpy import pi,cos,sin
import matplotlib.pyplot as plt
#https://ntrs.nasa.gov/api/citations/20205009350/downloads/03-KRUSTY%20Reactor%20Design.pdf
```

```python
krusty = openmc.Model()
```

```python
#materials
u8mo = openmc.Material(1,name = 'fuel')
u8mo.add_element('U',8,enrichment = 5.0)
u8mo.add_element('Mo',1)
u8mo.set_density('g/cm3',18.0)
water = openmc.Material(2,name = 'water')
water.add_element('H',2)
water.add_element('O',1)
water.set_density('g/cm3',1.0)
water.add_s_alpha_beta('c_H_in_H2O')
sodium = openmc.Material(3,name = 'sodium')
sodium.add_element('Na',1)
sodium.set_density('g/cm3',10.0)
nickel = openmc.Material(4,name = 'nickel')
nickel.add_element('Ni',1)
nickel.set_density('g/cm3',9.0)
mo = openmc.Material(5,name = 'molybdenum')
mo.add_element('Mo',1)
mo.set_density('g/cm3',10.0)
bc4 = openmc.Material(6,name = 'BC4')
#bc4.add_element('B',1)
bc4.add_element('C',4)
bc4.set_density('g/cm3',2.52)
krusty.materials = openmc.Materials([u8mo,water,sodium,nickel,mo,bc4])
```

```python
def hp_coord_find(n,num_hps):
    theta = n*np.pi*2/num_hps
    r = 11/2
    x = r*cos(theta)
    y = r*sin(theta)
    x = np.around(x,8)
    y = np.around(y,8)
    return x,y
def zcylinder(r,x,y):
    return openmc.ZCylinder(r = r,x0 = x, y0 = y)
    
hp_coords = [hp_coord_find(n,8) for n in range(8)]

# beginning geometry
heatpipe_surfs = [[zcylinder(1.27/2,x,y),zcylinder(1.1/2,x,y),zcylinder(1.05/2,x,y)] for x,y in hp_coords]
fuel_outer = openmc.ZCylinder(r=11.0/2,x0=0,y0=0)
fuel_inner = openmc.ZCylinder(r=4.0/2,x0=0,y0=0)
bottom = openmc.ZPlane(z0= -12.5, boundary_type = 'reflective')
top = openmc.ZPlane(z0= 12.5, boundary_type = 'reflective')
out = openmc.ZCylinder(r= 5.6+1.27/2,x0=0,y0=0, boundary_type = 'reflective')
#regions



chamber_cell = openmc.Cell(region = -fuel_inner & +bottom & -top,fill = bc4)
outside_region = -out&+fuel_outer & +bottom &-top
cells = [chamber_cell]
fuel_region = +bottom & -top &+fuel_inner & -fuel_outer
for hp,oni,ini in heatpipe_surfs:
    fuel_region &= +hp
    outside_region &= +hp
    cells.append(openmc.Cell(region = -hp &+oni &-top &+bottom,fill = sodium))
    cells.append(openmc.Cell(region = +ini &-oni&-top &+bottom,fill = nickel))
    cells.append(openmc.Cell(region = -ini&-top &+bottom,fill = None))
fuel_cell = openmc.Cell(region = fuel_region,fill = u8mo, name = 'fuel')
outside = openmc.Cell(region = outside_region,fill = mo, name = 'outter')
cells.append(fuel_cell)
cells.append(outside)
universe = openmc.Universe(cells = cells)
krusty.geometry = openmc.Geometry(universe)
universe.plot(width = (13,13),basis='xy', color_by= 'material')
```

```python
krusty.settings.source = openmc.IndependentSource(space=openmc.stats.Box(
    [-5.5, -5.5, -12.5], [5.5, 5.5, 12.5],only_fissionable=True))
krusty.settings.batches = 400
krusty.settings.inactive = 200
krusty.settings.particles = 1000

flux_tally_zernike = openmc.Tally()
flux_tally_zernike.scores = ['kappa-fission']
zernike_filter = openmc.ZernikeFilter(order=50, x=0.0, y=0.0, r=5.5)
cell_filter = openmc.CellFilter(fuel_cell)
flux_tally_zernike.filters = [cell_filter, zernike_filter]
krusty.tallies = openmc.Tallies([flux_tally_zernike])

krusty.export_to_xml()
```
